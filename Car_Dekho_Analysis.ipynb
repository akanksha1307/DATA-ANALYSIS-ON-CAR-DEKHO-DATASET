# **CASE STUDY : DATA ANALYSIS ON CAR DEKHO DATASET**

Here the dataset used is the Car Dekho dataset in the csv file. The data contain the 9 attribute


*   Car name: This column represents the name or model of the car.
*   Year: The manufacturing year of the car.

*   Selling Price: The price at which the car is being sold.
*   Present Price: The current market price or the price of the car when it was new.
*   Kms Driven: The total number of Kilometers the car has been driven.
*   Fuel Type: Indicates the type of fuel the car uses, such as Petrol, Diesel, or CNG.

*   Seller Type: Specifies whether the seller is an individual or a dealer.
*   Transmission: Indicates the type of transmission, whether Manual or Automatic.
*   Owner: The number of previous owners the car has had.


After the analysis we will be able to answer many question related to the dataset. The question can be :
What are the most common car models in the dataset?
1. How does the age of the car impact its selling price?
2. What is the distribution of selling prices in the dataset?
3. How does the distance driven (kms) affect the selling price?
4. What is the prevalence of different fuel types, and does it impact pricing?
5. How does the seller type (individual or dealer) influence selling prices?
6. What is the distribution of manual and automatic transmissions?
7. Does the number of previous owners impact the selling price?
